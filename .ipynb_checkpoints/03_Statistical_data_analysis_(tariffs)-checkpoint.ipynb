# Выбор лучшего тарифа

В нашем распоряжении имеются данные о 500-та клиентах федерального оператора сотовой связи компании «Мегалайн» по двум тарифам: тариф Smart и тариф Ultra. Для корректировки рекламного бюджета необходимо проанализировать поведение клиентов и выяснить, какой тариф приносит больше денег компании.

#### Необходимые функции

```python
def my_read(name):                                        # Загрузка датасетов
    try:
        df_read = pd.read_csv('/datasets/' + name)
    except:
        df_read = pd.read_csv(r'C:\Users\Admiral\Desktop\учеба\практикум\Проекты\Datasets\03_' + name)
    return df_read



def my_violinplot(df, col, title, ylable):                # Скрипичный график
    mpl.style.use(['seaborn-darkgrid'])
    plt.figure(figsize=(4, 5))
    sns.violinplot(y=df[col], color='#FF5555')
    plt.title(title)
    plt.ylabel(ylable)


def my_hist(dt, col, title, xlable, binns, x1, x2, x3):   # Гистограмма
    plt.figure(figsize=(6, 4))
    sns.distplot(dt[col], bins=binns)
    plt.title(title)
    plt.xlabel(xlable)
    plt.xticks(np.arange(x1, x2, x3))
    

def my_revenue(row):                                      # Подсчет выручки
    rev = row['month_fee']
    if row['calls_duration'] > row['minut_in']:
        rev += (row['calls_duration'] - row['minut_in']) * row['rub_minut']
    if row['msg_sum'] > row['msg_in']:
        rev += (row['msg_sum'] - row['msg_in']) * row['rub_msg']
    if row['mb_used'] > row['mb_in']:
        mb = row['mb_used'] - row['mb_in']
        gb = ceil(mb / 1024)
        rev += gb * row['rub_gb']
    return rev


def IQR1_5(df, col):                                      # Определение границ выброса
    Q1 = df[col].quantile(0.25)
    Q2 = df[col].quantile(0.75)
    IQR = Q2 - Q1
    min_col = Q1 - IQR * 1.5
    max_col = Q2 + IQR * 1.5
    if min_col < 0:
        print('Нижняя граница выброса:', 0)
        print('Верхняя граница выброса:', max_col)
    else:    
        print('Нижняя граница выброса:', min_col)
        print('Верхняя граница выброса:', max_col)
        
        

def analysis(df, usluga):                                # анализ услуги
    
    # статистики
    pivot = df.pivot_table(index = 'tariff', values = usluga,\
                           aggfunc = ['mean', 'median', 'var', 'std'])
    print('Статистики по ', usluga)
    display(pivot)
    
    # гистграмма
    fig = px.histogram(df, x = usluga, color = 'tariff',\
                       marginal = 'box', title = 'Распределение '+usluga)
    fig.show()
```

#### Объединение таблиц

Сохраним в переменную сгруппированную таблицу по ID и месяцам с количеством звонков и суммой минут по каждому месяцу и клиенту. Выведем первые 5 строк

```python
calls = df_calls.groupby(['user_id', 'month'])\
    .agg({'duration': 'sum', 'id': 'count'})\
    .reset_index()\
    .rename(columns= {'id': 'calls_count', 'duration': 'calls_duration'})
display(calls.head())
```

Сохраним в переменную сгруппированную таблицу по ID и месяцам с количеством отправленных сообщений по каждому месяцу и клиенту.

```python
msg = df_msg.groupby(['user_id', 'month'])\
    .agg({'id': 'count'})\
    .reset_index()\
    .rename(columns= {'id': 'msg_sum'})
display(msg.head())
```

Сохраним в переменную сгруппированную таблицу по ID и месяцам с суммой использованного трафика по каждому месяцу и клиенту.

```python
inet = df_inet.groupby(['user_id', 'month'])\
    .agg({'mb_used': 'sum'})\
    .reset_index()
display(inet.head())
```

```python
df_main[['month', 'rub_msg', 'rub_minut', 'calls_count', 'msg_sum', 'rub_gb']] =\
    df_main[['month', 'rub_msg', 'rub_minut', 'calls_count', 'msg_sum', 'rub_gb']].astype('uint8')
df_main[['user_id', 'calls_duration', 'msg_in', 'mb_in', 'minut_in', 'month_fee', 'mb_used']] =\
    df_main[['user_id', 'calls_duration', 'msg_in', 'mb_in', 'minut_in', 'month_fee', 'mb_used']]\
    .astype('uint16')
df_main['tariff'] = df_main['tariff'].astype('category')
```

#### Выдвижение гипотиз

##### Средняя выручка пользователей тарифов Ultra и Smart различается

Для отвержения/не отвержения гипотизы для начала сформулируем нулевую гипотизу так, чтобы в последствии рабочей гипотизой была альтернативная. Формулировка будет такая: Средняя выручка рассматриваемой выборки по тарифу Smart равна средней выручке рассматриваемой выборки тарифа Ultra. На основании нулевой гипотизы сформулируем двухстороннюю альтернативную гипотезу: Среднии выручки рассматриваемых выборок по тарифам Smart и Ultra различны. Чтобы задать критический уровень статистической значимости измерим длинны выборок по каждому тарифу. 

```python
display(len(df_fin.query('tariff == "smart"')), len(df_fin.query('tariff == "ultra"')))
```

Изучим дисперсии для обоих выборок. Так как выборки у нас не генеральные, методу "var" передадим параметр "ddof" с значением "1".

```python
print('Дисперсия выручки тарифа Smart: {:.0f}'\
      .format(df_fin.query('tariff == "smart"')['revenue'].var(ddof=1)))
```

```python
print('Дисперсия выручки тарифа Ultra: {:.0f}'\
      .format(df_fin.query('tariff == "ultra"')['revenue'].var(ddof=1)))
```

Объемы выборок больше 100 в пределах 2000. Принимаем критический уровень статистической значимости равный 0.01. Сравнивать среднии показатели выборок мы будем при помощи метода "scipy.stats.ttest_ind". В качестве параметров мы передадим массивы с выручкой по каждому тарифу, а так же в связи с тем, что дисперсии выборок различаются почти в 4 раза, мы передадим параметр "equal_var" со значенеием "False".

```python
p_tariff_revenue = st.ttest_ind(df_fin.query('tariff == "smart"')['revenue'], \
                   df_fin.query('tariff == "ultra"')['revenue'], equal_var=False)
alpha = 0.01
print('p-значение: {}'.format(p_tariff_revenue.pvalue))
if p_tariff_revenue.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

```python
display(len(df_fin.query('tariff == "smart" and revenue_clear != 0')), \
        len(df_fin.query('tariff == "ultra" and revenue_clear != 0')))
```

Изучим дисперсию выборок

```python
print('Дисперсия выручки тарифа Smart: {:.0f}'\
      .format(df_fin.query('tariff == "smart" and revenue_clear != 0')['revenue_clear'].var(ddof=1)))
```

```python
print('Дисперсия выручки тарифа Ultra: {:.0f}'\
      .format(df_fin.query('tariff == "ultra" and revenue_clear != 0')['revenue_clear'].var(ddof=1)))
```

Принимаем критический уровень статистической значимости равный 0.01. Методу "scipy.stats.ttest_ind" мы передадим  параметр "equal_var" со значенеием "False" так как различие между дисперсиями превышает 5%.

```python
p_tariff_revenue_clear = st.ttest_ind(df_fin.query('tariff == "smart" and revenue_clear != 0')\
                   ['revenue_clear'], df_fin\
                   .query('tariff == "ultra" and revenue_clear != 0')['revenue_clear'], \
                   equal_var=False)
alpha = 0.01
print('p-значение: {}'.format(p_tariff_revenue_clear.pvalue))
if p_tariff_revenue_clear.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

На основании полученного p-значения можно утверждать, что с вероятностью более 17% среднии выручки без учета ежемесячной платы по тарифу для обоих тарифов будут равны. Процент слишком велик чтобы отвергать нулевую гипотезу.

##### Cредняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов

Для отвержения/не отвержения гипотизы для начала сформулируем нулевую гипотизу так, чтобы в последствии рабочей гипотизой была альтернативная. Формулировка будет такая: Средняя выручка рассматриваемой выборки по Москве равна средней выручке рассматриваемой выборки остальных регионов. На основании нулевой гипотизы сформулируем двухстороннюю альтернативную гипотезу: Среднии выручки рассматриваемых выборок по Москве и остальным регионам различны. Чтобы задать критический уровень статистической значимости измерим длинны выборок по каждому тарифу.

```python
display(len(df_fin.query('city == "Москва"')), len(df_fin.query('city != "Москва"')))
```

Изучим дисперсию выборок

```python
print('Дисперсия выручки по Москве: {:.0f}'\
      .format(df_fin.query('city == "Москва"')['revenue'].var(ddof=1)))
```

```python
print('Дисперсия выручки по Москве: {:.0f}'\
      .format(df_fin.query('city != "Москва"')['revenue'].var(ddof=1)))
```

Принимаем критический уровень статистической значимости равный 0.01. Методу "scipy.stats.ttest_ind" мы передадим  параметр "equal_var" со значенеием "False", так как различие между дисперсиями превышает 5%.

```python
p_city_revenue = st.ttest_ind(df_fin.query('city == "Москва"')['revenue'], \
                   df_fin.query('city != "Москва"')['revenue'], equal_var=False)
alpha = 0.01
print('p-значение: {}'.format(p_city_revenue.pvalue))
if p_city_revenue.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

На основании p-значения можно сделать вывод, что с вероятностью в 50%  различие между средними выручками в Москве и остальными населенными пунктами  будет получено случайно. Принимаем нулевую гипотезу.

##### Средняя выручка пользователей тарифа Ultra больше средней выручки пользователей тарифа Smart

Для отвержения/не отвержения гипотизы для начала сформулируем нулевую гипотизу так, чтобы в последствии рабочей гипотизой была альтернативная. Формулировка будет такая: Средняя выручка рассматриваемой выборки по тарифу Smart равна средней выручке рассматриваемой выборки тарифа Ultra. На основании нулевой гипотизы сформулируем одностороннюю альтернативную гипотезу: Средняя выручка рассматриваемой выбороки по тарифу Ultra больше средней выручки пользователей тарифа Smart. По аналогии с первым тестом зададим критический уровень статистической значимости равный 0.01. Учитывая серьезное различие между дисперсиями выборок  методу "scipy.stats.ttest_ind" мы передадим параметр "equal_var" со значенеием "False". 

```python
p_tariff_revenue2 = st.ttest_ind(df_fin.query('tariff == "smart"')['revenue'],\
                                      df_fin.query('tariff == "ultra"')['revenue'],\
                                      equal_var=False)
alpha = 0.01
print('p-значение: {}'.format(p_tariff_revenue2.pvalue / 2))
if p_tariff_revenue2.pvalue / 2 < alpha and \
    df_fin.query('tariff == "smart"')['revenue'].mean() < \
    df_fin.query('tariff == "ultra"')['revenue'].mean():
        print("Отвергаем нулевую гипотезу")
else:
        print("Не получилось отвергнуть нулевую гипотезу")
```

С большой долей вероятности можно считать что в генеральной выборке средняя выручка от пользователей тарифа Ultra превышает среднюю выручку пользователей тарифа Smart.
